```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data_science_job.csv')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
cols = [var for var in df.columns if df[var].isnull().mean() < 0.05 and df[var].isnull().mean() > 0]

cols
```

```python
df[cols].sample(5)
```

```python
df['education_level'].value_counts()
```

```python
df['enrolled_university'].value_counts()
```

```python
len(df[cols].dropna()) / len(df)
```

```python
new_df = df[cols].dropna()

df.shape, new_df.shape
```

```python
new_df.hist(bins=50,density=True,figsize=(12,12))
```

```python
df.hist(bins=20,density=True,figsize=(12,12))
```

```python
df.head()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

new_df['training_hours'].hist(bins=50,ax=ax,density=True,color='green')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

# Original Data

df['training_hours'].hist(bins=50,ax=ax, density=True, color='red')

```

```python
new_df['training_hours']
```

```python
# data after cca, the argument alpha makes the color transparent, so we can
# see the overlay of the 2 distributions
fig = plt.figure()
ax = fig.add_subplot(111)
df['training_hours'].hist(bins=50,ax=ax, density=True, color='red')

new_df['training_hours'].hist(bins=50,ax=ax,color='green',density=True,alpha=0.)
```

```python
fig = plt.figure()
ax = fig.add_subplot()

df['training_hours'].plot.density(color='red')

new_df['training_hours'].plot.density(color='green')
```

```python
fig = plt.figure()

ax = fig.add_subplot()

df['city_development_index'].hist(bins=50,density=True,ax=ax,color='red')


# data after cca, the argument alpha makes the color transparent, sow we can
# see the overlay of the 2 distributions.

new_df['city_development_index'].hist(bins=50,density=True,ax=ax,color='green',alpha=0.8)
```

```python
fig = plt.figure()
ax = fig.add_subplot()

df['city_development_index'].plot.density(color='red')
new_df['city_development_index'].plot.density(color='green')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

df['experience'].hist(bins=50,ax=ax,density=True,color='red')

new_df['experience'].hist(bins=50,ax=ax,density=True,color='green',alpha=0.8)
```

```python
fig = plt.figure()
ax = fig.add_subplot()


df['experience'].plot.density(color='red')
new_df['experience'].plot.density(color='green')
```
