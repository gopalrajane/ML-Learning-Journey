#  Multivariate Analysis.

```python
import pandas as pd
import seaborn as sns
```

<h3>1. Scatterplot (Numerical - Numerical)</h3>

```python
tips = sns.load_dataset('tips')
tips.sample(5)
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

```python
sns.scatterplot(x='total_bill',y='tip',hue='sex',data=tips)
```

```python
sns.scatterplot(x='total_bill',y='tip',hue='sex',style='smoker',data=tips)
```

<h3>2. Barplot (Numerical - Categorical)</h3>


```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
sns.barplot(x='pclass',y='fare',data=df)
```

```python
sns.barplot(x='sex',y='survived',data=df)
```

```python
sns.barplot(x='pclass', y='age', hue='sex',data=df)
```

<h3>3. Boxplot (Numerical - Categorical)

```python
sns.boxplot(x='sex',y='age',data=df)
```

```python
sns.boxplot(x='sex',y='age',hue='survived',data=df)
```

<h3>4. Distplot (Numerical - Categorical)

```python
sns.histplot(x='age',data=df)
```

```python
sns.histplot(x='age',hue='survived',data=df)
```

<h3>5. Kdeplot 

```python
sns.kdeplot(x='age',hue='survived',data=df)
```

```python
df.sample(5)
```

```python
sns.kdeplot(x='pclass',hue='survived',data=df)
```

<h3>6. Heatmap (Categorical - Categorical)</h3>


```python
pd.crosstab(df['pclass'],df['survived'])
```

```python
sns.heatmap(pd.crosstab(df['pclass'],df['survived']))
```

```python
df.groupby('pclass')['survived'].sum()*100
```

```python
(df.groupby('pclass')['survived'].mean()*100).plot(kind='bar')
```

```python
sns.pairplot(tips)
```
