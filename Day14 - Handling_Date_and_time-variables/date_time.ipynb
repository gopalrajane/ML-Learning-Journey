# Working with Dates

```python
# Converting the column date into datetime datatype

date['date'] = pd.to_datetime(date['date'])
```

```python
date['date'].info()
```

<h4 style="color:green">1. Extract Year</h4>

```python
date['date_year'] = date['date'].dt.year
```

```python
date['date_year'].head()
```


<h4 style="color:green">2. Extract Month</h4>


```python
date['date_month'] = date['date'].dt.month
```

```python
date['date_month'].head()
```

<h4 style="color:green">3. Extract Month Name</h4>


```python
date['date_month_name'] = date['date'].dt.month_name()
```

```python
date['date_month_name']
```

```python
date.head()
```

<h4 style="color:green">4. Extract Day</h4>


```python
date['date_day'] = date['date'].dt.day
```

```python
date['date_day']
```

```python
date.head()
```

<h4 style="color:green">5. Extract day of week</h4>


```python
date['date_dow'] = date['date'].dt.day_of_week
```

```python
date['date_dow']
```

<h4 style="color:green">6. Day of week-name</h4>


```python
date['day_dow_name'] = date['date'].dt.day_name()
```

```python
date['day_dow_name'].head()
```

```python
date.drop(columns=['product_id','city_id','orders'])
```

```python
date.head()
```

<h4 style="color:green">7. is Weekend on day ?</h4>


```python
date['date_is_weekend'] = np.where(date['day_dow_name'].isin(['Sunday','Saturday']),1,0)

date.head()
```

<h4 style="color:green">8. Extract week of the year</h4>



```python
date['date_week'] = date['date'].dt.weekday
```

```python
date.head()
```

<h4 style="color:green">9. Extract Quater</h4>


```python
date['date_quater'] = date['date'].dt.quarter
```

```python
date.head()
```

<h4 style="color:green">10. Extract Semester</h4>


```python
date['date_semester'] = np.where(date['date_quater'].isin([1,2]),1,2)

date.head()
```
